```python
pd.to_datetime(t['Date']) # Changing the data type of the column for smooth Graphical Representation
```

```python
# Plotting the graph

plt.figure(figsize=(12, 6)) # Define the figure size you want
plt.plot(t['Date'], t['Close'], label='Closing Price')

# Customize the plot
plt.title('Stock Price')
plt.xlabel('Date')
plt.ylabel('Close')
plt.legend()
plt.grid(False) # Used for the griding effeect in the chart

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(110, 16))
plt.bar(t["Date"], t['Volume'], label='Volume')

# Customize the plot
plt.title("Stock Volume")
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```
